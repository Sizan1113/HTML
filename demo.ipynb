```python
# Write code below ðŸ’–
Gryffindor = 0
Ravenclaw = 0
Hufflepuff = 0
Slytherin = 0
print("Do you like Down or Dusk?")
print("1) Dawn")
print("2)Dusk")
ans = int(input("Enter your answer"))
if ans == 1:
  Gryffindor += 1
  Ravenclaw +=1
elif ans == 2:
  Hufflepuff +=1
  Slytherin +=1
else:
  print("Please enter 1 or 2.")
print("When I'm dead,I want people to remember me as:")
print("1)The Good")
print("1)The great")
print("1)The wise")
print("1)The bold")
ans = int(input("Enter your answer from 1 to 4"))
if ans == 1:
  Hufflepuff +=2
elif ans == 2:
  Slytherin +=2
elif ans == 3:
  Ravenclaw +=2
elif ans == 4:
  Gryffindor +=2
else:
  print("Please enter number from 1 - 4")
print(" Which kind of instrument most pleases your ear?")
print("1)The violin")
print("2)The trumpet")
print("3)The piano")
print("4)The drum")
ans = int(input("Enter number from 1 - 4:"))
if ans == 1:
  Slytherin +=4
elif ans == 2:
  Hufflepuff +=4
elif ans == 3:
  Ravenclaw +=4
elif ans == 4:
  Gryffindor +=4
else:
  print("Please enter number from 1 - 4")
print(f"The score for house Gryffindor:",Gryffindor)
print(f"The score for house Ravenclaw:",Ravenclaw)
print(f"The score for house Hufflepuff:",Hufflepuff)
print(f"The score for house Slytherin:",Slytherin)
if Gryffindor > Hufflepuff and Hufflepuff > Ravenclaw and Ravenclaw > Slytherin:
  print(f"The house with the highest score is:",Gryffindor)
elif Hufflepuff > Ravenclaw and Ravenclaw > Slytherin and Slytherin > Gryffindor:
  print(f"The house with the highest score is:",Hufflepuff)
elif  Ravenclaw > Slytherin and Slytherin > Gryffindor and Gryffindor > Hufflepuff:
  print(f"The house with the highest score is:",Ravenclaw)
else:
  print(f"The house with the highest score is:",Slytherin)
```
